### Réponse 02:
Formule 01:(Quantité exprimé en milliers de tonnes)

Disponibilité intérieure= production + importation + variation stock - exportation.

Formule 02:(Quantité exprimé en milliers de tonnes)

Disponibilité intérieure= aliments pour animaux + semences + perte + traitement + autres utilisation (non alimentaire) + nourriture.

#### Illustration avec l'exemple du blé pour la france

```python
df_veg=pd.DataFrame(pd.read_csv("fr_vegetaux.csv"))#lire la base de donnée 
fr_bl=df_veg[(df_veg['Zone']=="France")&(df_veg['Produit']=="Blé")]#étudier la base de données avec le blé de la france (que m'intéresse)
di=fr_bl[(fr_bl['Élément']=="Disponibilité intérieure")].Valeur.sum()#sélectionner la valeur de la disponibilité intèrieur
print('Disponiblité intèrieur est:' ,di)# afficher la valeur de la disponibilité intèrieur
```

```python
pr=fr_bl[fr_bl['Élément']=="Production"]['Valeur'].sum()
im=fr_bl[fr_bl['Élément']=="Importations - Quantité"].Valeur.sum()#j'ai mis la fonction sum au lieu de values juste pour l'affichage
var=fr_bl[fr_bl['Élément']=="Variation de stock"].Valeur.sum()
ex=fr_bl[(fr_bl['Élément']=="Exportations - Quantité")].Valeur.sum()
print('Disponiblité intèrieur calculée à partir la première formule est:' ,im+var+pr-ex )
```

```python
pe=fr_bl[(fr_bl['Élément']=="Pertes")].Valeur.sum()
tr=fr_bl[(fr_bl['Élément']=="Traitement")].Valeur.sum()
nr=fr_bl[(fr_bl['Élément']=="Nourriture")].Valeur.sum()
al_ani=fr_bl[(fr_bl['Élément']=="Aliments pour animaux")].Valeur.sum()
aut_ut=fr_bl[(fr_bl['Élément']=="Autres utilisations (non alimentaire)")].Valeur.sum()
sm=fr_bl[(fr_bl['Élément']=="Semences")].Valeur.sum()
print('Disponiblité intèrieur calculée à partir la deuxième formule est:',al_ani+sm+pe+tr+aut_ut+nr)
```

### Réponse 03:

```python
df_ani=pd.DataFrame(pd.read_csv("fr_animaux.csv")) #lire la base de donnée 
df_ani_veg=pd.concat([df_ani, df_veg], ignore_index=True) # faire concatiner les deux tableaux fr_ani fr_veg pour avoir tous les produits dans un seul tab
df_ani_veg.rename(columns={'Valeur':'Valeur_produit'}, inplace=True)# renommer Valeur pour non pas confondus entre les deux colonnes qd on fusionne les 2 tab 
d.rename(columns={'Valeur':'Valeur_population'}, inplace=True) #             /////////////         /////////////
#étudier d au lieu df car d il contient pas la chine 
ddf=d[['Zone', 'Valeur_population']]#créer d'autre tab à partir d pour ne pas confondus entre les titres des colonnes élément les produits ... etc et prendre juste ce qu'on a besoin 
df_pop_ani_veg=pd.merge(ddf,df_ani_veg,on='Zone')# fusionner les 2 tab pour faire les calculs
df_pop_ani_veg_disp=df_pop_ani_veg[df_pop_ani_veg['Élément']=="Disponibilité alimentaire (Kcal/personne/jour)"].copy()#sélectionner les valeurs de la disp alimentaire pour chaque pays et chaque produit 
df_pop_ani_veg['Disponibilité alimentaire en kcal']=df_pop_ani_veg_disp['Valeur_population']*df_pop_ani_veg_disp['Valeur_produit']*365*1000 #ajouter une colonne de la disp alimentaire en kcal  
df_disp=df_pop_ani_veg.loc[df_pop_ani_veg['Élément']=="Disponibilité alimentaire (Kcal/personne/jour)",['Zone','Produit','Valeur_produit','Disponibilité alimentaire en kcal']]#séléctionner les colonnes qu'on veut afficher 
df_disp.rename(columns={'Valeur_produit':'Disponibilité alimentaire (Kcal/personne/jour)'}, inplace=True)
df_disp
```

```python
df_pop_ani_veg_pro=df_pop_ani_veg[(df_pop_ani_veg['Élément']=="Disponibilité de protéines en quantité (g/personne/jour)")]#sélectionner les valeurs de la disp alimentaire de protéines pour chaque pays et chaque produit 
df_pop_ani_veg['Disponibilité alimentaire en kg de protéines']=df_pop_ani_veg_pro['Valeur_population']*df_pop_ani_veg_pro['Valeur_produit']*365*1000*0.001#ajouter une colonne de la disp alimentaire de protéines en kg  
df_pro=df_pop_ani_veg.loc[df_pop_ani_veg['Élément']=="Disponibilité de protéines en quantité (g/personne/jour)",['Zone','Produit','Valeur_produit','Disponibilité alimentaire en kg de protéines']] 
df_pro.rename(columns={'Valeur_produit':'Disponibilité de protéines en quantité (g/personne/jour)'}, inplace=True)
df_pro
```

### Réponse 04:

```python
 df_pop_ani_veg_nou=df_pop_ani_veg.loc[(df_pop_ani_veg['Élément']=="Nourriture")&(df_pop_ani_veg['Valeur_produit']!=0),['Zone','Produit','Valeur_produit']]#sélectionner les valeurs de la nourriture non nuls pour chaque pays et chaque produit (pour éviter la division sur 0)
df_ratio_kcal=df_disp.merge(df_pop_ani_veg_nou,on=['Zone','Produit'])# fusionner les 2 tab pour faire les calculs
df_ratio_kcal['Ratio (kcal/kg)']=df_ratio_kcal['Disponibilité alimentaire en kcal']/(df_ratio_kcal['Valeur_produit']*1000000)# multiplier la valeur de produit par 1000000 pour faire une transformation d'unité(milliers de tonne-> kg)
df_pop_ani_veg['Ratio (kcal/kg)']=df_ratio_kcal['Ratio (kcal/kg)']# ajouter une colonne Ratio (kcal/kg)
df_ratio_kcal
```

```python
print(df_ratio_kcal.loc[df_ratio_kcal['Produit']=="Oeufs",['Ratio (kcal/kg)']].mean().sum(),'kcal/kg')# exemple de vérification sur le site la valeur calorique 147 kcal pour 100g donc 1470 kcal/kg 
```

```python
df_ratio_kg=df_pro.merge(df_pop_ani_veg_nou,on=['Zone','Produit'])# fusionner les 2 tab pour faire les calculs
df_ratio_kg['Ratio (kg/kg)']=df_ratio_kg['Disponibilité alimentaire en kg de protéines']*100/(df_ratio_kg['Valeur_produit']*1000000)#*100 calculer le pourcentage de protéines
df_ratio_kg[['Zone','Produit','Ratio (kg/kg)']]
df_pop_ani_veg['Ratio (kg/kg)']=df_ratio_kg['Ratio (kg/kg)']# ajouter une colonne Ratio (kg/kg)
df_ratio_kg
```

```python
 # exemple de vérification selon le wikipedia la valeur est 10.7g pour 100g donc 0.107 kg/kg (10.7%)
print(df_ratio_kg.loc[(df_ratio_kg['Produit']=="Avoine"),['Ratio (kg/kg)']].mean().sum(),'%')# calcluer la moyenne  
```

### Réponse 05:

```python
pr_df_ratio_kcal=df_ratio_kcal.groupby("Produit")['Ratio (kcal/kg)'].mean()#Calculer la moyenne de la ratio(kcal/kg) de chaque produit
pr_df_ratio_kg=df_ratio_kg.groupby("Produit")['Ratio (kg/kg)'].mean()#Calculer la moyenne de la ratio(kg/kg) de chaque produit
ratio=pd.DataFrame({"ratio_kal/kg":pr_df_ratio_kcal, "ratio_kg/kg":pr_df_ratio_kg})#Créer une nouvelle table
ratio.sort_values("ratio_kal/kg", ascending=False).head(5)# classer les valeurs en ordre décroissant et afficher les cinq premiers
```

```python
ratio.sort_values("ratio_kg/kg", ascending=False)[:5]# classer les valeurs du plus grand au plus petit et afficher les cind premiers
```

### Réponse 06:

```python
df_veg_pop=ddf.merge(df_veg, on='Zone')# fusionner les 2 tab df_veg et ddf (pour ne pas prendre la chine en considération)
df_veg_dis=df_veg_pop.loc[df_veg_pop['Élément']=="Disponibilité intérieure",['Produit','Valeur']]#sélectionner les valeurs de la disp intérieur pour chaque pays et chaque produit végétaux 
df_veg_dis['disponibilité intérieure(kg)']=df_veg_dis['Valeur']*1000000 # calculer la disponibilité intérieur en kg
disp_kg=df_veg_dis.groupby('Produit')['disponibilité intérieure(kg)'].sum()# calculer la somme de la disponibilité int de chaque produit 
veget=ratio.merge(disp_kg,left_index=True, right_index=True)#utiliser l'index comme clé de jointure pour la fusion des DataFrames
veget['disponibilité intérieure(kcal)']=veget['ratio_kal/kg']*veget['disponibilité intérieure(kg)']# calculer la disponibilité intèrieure en kcal par produit 
disp_int_mondiale=veget['disponibilité intérieure(kcal)'].sum()#calculer la somme des kilo calories  
print("la disponibilité intérieure mondiale pour les produits végétaux exprimée en kcal est:",disp_int_mondiale)
```

### Réponse 07:

```python
humaine_nourr=disp_int_mondiale/(365*2500)# une personne a besoin 2500 kcal par jour
print("Nombre de personnes qui peuvent être nourries si toute la disponibilité intérieure mondiale des produits végétaux  était utilisée pour la nourriture en terme de calories est: ",humaine_nourr)
print("en pourcentage",humaine_nourr/pop*100,'%')# diviser par le nombre total de la population
```

```python
veget['disp_int_protéines(kg)']=veget['ratio_kg/kg']*veget['disponibilité intérieure(kg)']*0.01 # calculer la disp int en terme de protéines par produit (multipluier par 0.01 psq on a les valeurs en pourcentage)
disp_int_kg_mondiale=veget['disp_int_protéines(kg)'].sum()# calculer la somme total
humaine_nourr_pro=disp_int_kg_mondiale/(365*0.060)# une personne a besoin 60 g par jour
print("Nombre de personnes qui peuvent être nourries si toute la disponibilité intérieure mondiale des produits végétaux  était utilisée pour la nourriture en terme de protéines est: ",humaine_nourr_pro)
print("en pourcentage",humaine_nourr_pro/pop*100,'%')# diviser par le nombre total de la population
```

### Réponse 08:

```python
df_veg_pert=df_veg_pop.loc[(df_veg_pop['Élément']=="Pertes")|(df_veg_pop['Élément']=="Aliments pour animaux")|(df_veg_pop['Élément']=="Disponibilité intérieure"),['Produit','Valeur']]#sélectionner les valeurs (pertes, aliments pour animaux, disp int) pour chaque pays et chaque produit végétaux
df_veg_pert['disp_int/pertes/ali_animaux(kg)']=df_veg_pert['Valeur']*1000000 # la quantité en kg
per_ali=df_veg_pert.groupby('Produit')['disp_int/pertes/ali_animaux(kg)'].sum()#calculer la somme pour chaque produit 
veget_per=ratio.merge(per_ali,left_index=True, right_index=True)#utiliser l'index comme clé de jointure pour la fusion des DataFrames
veget_per['disp_int/pertes/ali_animaux(kcal)']=veget_per['ratio_kal/kg']*veget_per['disp_int/pertes/ali_animaux(kg)']#calculer la quantité en terme d'energie 
per_ali_kcal_mondiale=veget_per['disp_int/pertes/ali_animaux(kcal)'].sum()#calculer la somme des kilo calories  
print("énergie totale en kcal est:",per_ali_kcal_mondiale)
veget_per['disp_int/pertes/ali_animaux(kg)']=veget_per['ratio_kg/kg']*veget_per['disp_int/pertes/ali_animaux(kg)']*0.01 #calculer la quantité en terme de protéines par produit
per_ali_kg_mondiale=veget_per['disp_int/pertes/ali_animaux(kg)'].sum()
print("protéines totales en kg est:",per_ali_kg_mondiale)
```

```python
humaine_nou=per_ali_kcal_mondiale/(365*2500)# une personne a besoin 2500 kcal par jour
print("Nombre de personnes qui peuvent être nourries en terme de calories est:",humaine_nou)
print("en pourcentage",humaine_nou/pop*100,'%')# diviser par le nombre total de la population
```

```python
humaine_nou_pro=per_ali_kg_mondiale/(365*0.060)# une personne a besoin 60 g par jour
print("Nombre de personnes qui peuvent être nourries en terme de protéines est:",humaine_nou_pro)
print("en pourcentage",humaine_nou_pro/pop*100,'%')# diviser par le nombre total de la population
```

```python
pay_sous_ali=df_sous_ali_2013.loc[(df_sous_ali_2013['Valeur']>0),['Zone']]#sélectionner les pays dans lesquels la FAO recense des personnes en sous-nutrition en 2013
df_pay_souali=df_pop_ani_veg.merge(pay_sous_ali,on=['Zone'])# fusionner les 02 tab pour obtenir plus d'informations sur les pays souffrant de sous-nutrition
df_pay_souali=df_pay_souali.loc[df_pay_souali['Élément']=="Exportations - Quantité",['Zone','Produit','Valeur_produit']]#sélectionner les valeurs de l'exportations
df_pay_souali.rename(columns={'Valeur_produit':'Exportations'}, inplace=True)# renommer Valeur_produit par exportation
df_pro_pay=df_pay_souali.groupby('Produit')['Exportations'].sum().reset_index()# calculer la somme des valeurs de chaque produit 
df_pro_pay=df_pro_pay.sort_values( 'Exportations',ascending=False)[:15]# classer les valeurs du plus grand au plus petit et afficher les quinze premiers pays
df_pro_pay
```

```python
df_merg_bilan_pro=df_pop_ani_veg.merge(df_pro_pay,on=['Produit'])#fusionner les deux tab pour obtenir tous les info concernant les quinze produits les plus exportés par les pays dans lesquels la FAO recense des personnes en sous-nutrition en 2013 
piv1=df_merg_bilan_pro.pivot_table(index=['Zone','Produit'],columns='Élément',values='Valeur_produit')#utiliser des éléments présentes dans colonne "Élément" comme noms de colonnes 
piv1.reset_index(inplace=True)# réintégrer les lignes de l'index comme colonnes simples
df_p=piv1.sort_values('Importations - Quantité', ascending=False)[:200]# classer les valeurs de l'importation en ordre croissant et afficher les 200 premiers
df_p[['Zone','Produit','Importations - Quantité']]
```

```python
df_pr=piv1.groupby('Produit')[['Importations - Quantité','Autres utilisations (non alimentaire)','Disponibilité intérieure','Aliments pour animaux','Nourriture']].sum().reset_index()# calculer la somme (importation, autre utilisations, disp int, alim pour anim, nourriture) de chaque produit
df_pr['Ratio_AutreUti_DispInt']=df_pr['Autres utilisations (non alimentaire)']/df_pr['Disponibilité intérieure']#calculer le ratio entre la quantité destinés aux "Autres utilisations" et la disponibilité intérieure.
df_pr['Ratio_animaux_nourriture']=df_pr['Aliments pour animaux']/(df_pr['Nourriture']+df_pr['Aliments pour animaux'])#calculer le ratio entre la quantité destinée à la nourriture animale et la quantité destinée à la nourriture (animale + humaine)
df_pr
```

```python
df_pr.sort_values('Ratio_AutreUti_DispInt', ascending=False)[:3]#calculer les trois produits ayant la plus grande valeur pour le ratio entre la quantité destinée aux "autres utilisations" et la disponibilité intérieur
```
